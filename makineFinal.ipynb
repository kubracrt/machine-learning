```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, roc_auc_score, roc_curve, classification_report
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB

# Verisetini yükle
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
df = pd.read_csv(url, names=columns)

# Verisetini 70% eğitim ve 30% test olarak böl
X_veri = df.drop('Outcome', axis=1)
y_veri = df['Outcome']
X_egitim, X_test, y_egitim, y_test = train_test_split(X_veri, y_veri, test_size=0.3, random_state=42)

# Özellik ölçekleme
olcekleyici = StandardScaler()
X_egitim_olceklendi = olcekleyici.fit_transform(X_egitim)
X_test_olceklendi = olcekleyici.transform(X_test)

# Naive Bayes sınıflandırıcısı uygulama
naive_bayes = GaussianNB()
naive_bayes.fit(X_egitim_olceklendi, y_egitim)
y_tahmin_nb = naive_bayes.predict(X_test_olceklendi)

# Karışıklık matrisi ve sınıflandırma raporu
karisik_matrisi_nb = confusion_matrix(y_test, y_tahmin_nb)
sinif_raporu_nb = classification_report(y_test, y_tahmin_nb)
dogruluk_nb = accuracy_score(y_test, y_tahmin_nb)
f1_nb = f1_score(y_test, y_tahmin_nb)
roc_auc_nb = roc_auc_score(y_test, y_tahmin_nb)

# Sonuçları yazdır
print("Naive Bayes Sınıflandırma Sonuçları")
print("Karışıklık Matrisi:")
print(karisik_matrisi_nb)
print("\nSınıflandırma Raporu:")
print(sinif_raporu_nb)
print(f"Doğruluk: {dogruluk_nb:.2f}")
print(f"F1 Skoru: {f1_nb:.2f}")
print(f"ROC AUC Skoru: {roc_auc_nb:.2f}")

# ROC Eğrisi
fpr_nb, tpr_nb, _ = roc_curve(y_test, naive_bayes.predict_proba(X_test_olceklendi)[:, 1])
plt.figure()
plt.plot(fpr_nb, tpr_nb, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_nb)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Yanlış Pozitif Oranı')
plt.ylabel('Doğru Pozitif Oranı')
plt.title('ROC Eğrisi - Naive Bayes')
plt.legend(loc="lower right")
plt.show()
```

```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, roc_auc_score, roc_curve, classification_report
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Verisetini yükle
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
df = pd.read_csv(url, names=columns)

# Verisetini 70% eğitim ve 30% test olarak böl
X_veri = df.drop('Outcome', axis=1)
y_veri = df['Outcome']
X_egitim, X_test, y_egitim, y_test = train_test_split(X_veri, y_veri, test_size=0.3, random_state=42)

# Özellik ölçekleme
olcekleyici = StandardScaler()
X_egitim_olceklendi = olcekleyici.fit_transform(X_egitim)
X_test_olceklendi = olcekleyici.transform(X_test)

# En iyi k değerini bulma fonksiyonu
def en_iyi_k_bul(X_egitim, y_egitim, X_test, y_test):
    dogruluklar = []
    for k in range(1, 21):
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_egitim, y_egitim)
        y_tahmin = knn.predict(X_test)
        dogruluk = accuracy_score(y_test, y_tahmin)
        dogruluklar.append(dogruluk)
    en_iyi_k = dogruluklar.index(max(dogruluklar)) + 1
    return en_iyi_k

# En iyi k değerini belirle
en_iyi_k = en_iyi_k_bul(X_egitim_olceklendi, y_egitim, X_test_olceklendi, y_test)

# En iyi k değeri ile KNN sınıflandırıcısını uygulama
knn = KNeighborsClassifier(n_neighbors=en_iyi_k)
knn.fit(X_egitim_olceklendi, y_egitim)
y_tahmin_knn = knn.predict(X_test_olceklendi)

# Karışıklık matrisi ve sınıflandırma raporu
karisik_matrisi_knn = confusion_matrix(y_test, y_tahmin_knn)
sinif_raporu_knn = classification_report(y_test, y_tahmin_knn)
dogruluk_knn = accuracy_score(y_test, y_tahmin_knn)
f1_knn = f1_score(y_test, y_tahmin_knn)
roc_auc_knn = roc_auc_score(y_test, y_tahmin_knn)

# Sonuçları yazdır
print(f"\nEn Yakın Komşular Sınıflandırma Sonuçları (k={en_iyi_k})")
print("Karışıklık Matrisi:")
print(karisik_matrisi_knn)
print("\nSınıflandırma Raporu:")
print(sinif_raporu_knn)
print(f"Doğruluk: {dogruluk_knn:.2f}")
print(f"F1 Skoru: {f1_knn:.2f}")
print(f"ROC AUC Skoru: {roc_auc_knn:.2f}")

# ROC Eğrisi
fpr_knn, tpr_knn, _ = roc_curve(y_test, knn.predict_proba(X_test_olceklendi)[:, 1])
plt.figure()
plt.plot(fpr_knn, tpr_knn, color='darkorange', lw=2, label='ROC eğrisi (alan = %0.2f)' % roc_auc_knn)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Yanlış Pozitif Oranı')
plt.ylabel('Doğru Pozitif Oranı')
plt.title('ROC Eğrisi - KNN')
plt.legend(loc="lower right")
plt.show()
```

```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, roc_auc_score, roc_curve, classification_report
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC

# Verisetini yükle
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
df = pd.read_csv(url, names=columns)

# Verisetini 70% eğitim ve 30% test olarak böl
X_veri = df.drop('Outcome', axis=1)
y_veri = df['Outcome']
X_egitim, X_test, y_egitim, y_test = train_test_split(X_veri, y_veri, test_size=0.3, random_state=42)

# Özellik ölçekleme
olcekleyici = StandardScaler()
X_egitim_olceklendi = olcekleyici.fit_transform(X_egitim)
X_test_olceklendi = olcekleyici.transform(X_test)

# MLP sınıflandırıcısını uygulama
mlp_siniflandirici = MLPClassifier(max_iter=500, random_state=42)
mlp_siniflandirici.fit(X_egitim_olceklendi, y_egitim)
y_tahmin_mlp = mlp_siniflandirici.predict(X_test_olceklendi)

# Karışıklık matrisi ve sınıflandırma raporu
karisik_matrisi_mlp = confusion_matrix(y_test, y_tahmin_mlp)
sinif_raporu_mlp = classification_report(y_test, y_tahmin_mlp)
dogruluk_mlp = accuracy_score(y_test, y_tahmin_mlp)
f1_mlp = f1_score(y_test, y_tahmin_mlp)
roc_auc_mlp = roc_auc_score(y_test, y_tahmin_mlp)

# Sonuçları yazdır
print("\nMulti-Layer Perceptron Sınıflandırma Sonuçları")
print("Karışıklık Matrisi:")
print(karisik_matrisi_mlp)
print("\nSınıflandırma Raporu:")
print(sinif_raporu_mlp)
print(f"Doğruluk: {dogruluk_mlp:.2f}")
print(f"F1 Skoru: {f1_mlp:.2f}")
print(f"ROC AUC Skoru: {roc_auc_mlp:.2f}")

# ROC Eğrisi
fpr_mlp, tpr_mlp, _ = roc_curve(y_test, mlp_siniflandirici.predict_proba(X_test_olceklendi)[:, 1])
plt.figure()
plt.plot(fpr_mlp, tpr_mlp, color='darkorange', lw=2, label='ROC eğrisi (alan = %0.2f)' % roc_auc_mlp)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Yanlış Pozitif Oranı')
plt.ylabel('Doğru Pozitif Oranı')
plt.title('ROC Eğrisi - MLP')
plt.legend(loc="lower right")
plt.show()

# SVM sınıflandırıcısını uygulama
svm_siniflandirici = SVC(probability=True, random_state=42)
svm_siniflandirici.fit(X_egitim_olceklendi, y_egitim)
y_tahmin_svm = svm_siniflandirici.predict(X_test_olceklendi)

# Karışıklık matrisi ve sınıflandırma raporu
karisik_matrisi_svm = confusion_matrix(y_test, y_tahmin_svm)
sinif_raporu_svm = classification_report(y_test, y_tahmin_svm)
dogruluk_svm = accuracy_score(y_test, y_tahmin_svm)
f1_svm = f1_score(y_test, y_tahmin_svm)
roc_auc_svm = roc_auc_score(y_test, y_tahmin_svm)

# Sonuçları yazdır
print("\nDestek Vektör Makineleri Sınıflandırma Sonuçları")
print("Karışıklık Matrisi:")
print(karisik_matrisi_svm)
print("\nSınıflandırma Raporu:")
print(sinif_raporu_svm)
print(f"Doğruluk: {dogruluk_svm:.2f}")
print(f"F1 Skoru: {f1_svm:.2f}")
print(f"ROC AUC Skoru: {roc_auc_svm:.2f}")

# ROC Eğrisi
fpr_svm, tpr_svm, _ = roc_curve(y_test, svm_siniflandirici.predict_proba(X_test_olceklendi)[:, 1])
plt.figure()
plt.plot(fpr_svm, tpr_svm, color='darkorange', lw=2, label='ROC eğrisi (alan = %0.2f)' % roc_auc_svm)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Yanlış Pozitif Oranı')
plt.ylabel('Doğru Pozitif Oranı')
plt.title('ROC Eğrisi - SVM')
plt.legend(loc="lower right")
plt.show()
```
